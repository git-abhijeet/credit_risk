```python
# Remove nulls
df1 = df1.loc[df1['Age_Oldest_TL'] != -99999]
```

```python
columns_to_be_removed = []

for i in df2.columns:
    if df2.loc[df2[i] == -99999].shape[0] > 10000:
        columns_to_be_removed .append(i)



df2 = df2.drop(columns_to_be_removed, axis =1)



for i in df2.columns:
    df2 = df2.loc[ df2[i] != -99999 ]
```

```python
# Checking common column names
for i in list(df1.columns):
    if i in list(df2.columns):
        print (i)
```

```python
# Merge the two dataframes, inner join so that no nulls are present
df = pd. merge ( df1, df2, how ='inner', left_on = ['PROSPECTID'], right_on = ['PROSPECTID'] )

```

```python
# check how many columns are categorical
for i in df.columns:
    if df[i].dtype == 'object':
        print(i)
```

```python
# Chi-square test
for i in ['MARITALSTATUS', 'EDUCATION', 'GENDER', 'last_prod_enq2', 'first_prod_enq2']:
    chi2, pval, _, _ = chi2_contingency(pd.crosstab(df[i], df['Approved_Flag']))
    print(i, '---', pval)
```

```python
# VIF for numerical columns
numeric_columns = []
for i in df.columns:
    if df[i].dtype != 'object' and i not in ['PROSPECTID','Approved_Flag']:
        numeric_columns.append(i)
```

```python
# VIF sequentially check

vif_data = df[numeric_columns]
total_columns = vif_data.shape[1]
columns_to_be_kept = []
column_index = 0



for i in range (0,total_columns):
    
    vif_value = variance_inflation_factor(vif_data, column_index)
    print (column_index,'---',vif_value)
    
    
    if vif_value <= 6:
        columns_to_be_kept.append( numeric_columns[i] )
        column_index = column_index+1
    
    else:
        vif_data = vif_data.drop([ numeric_columns[i] ] , axis=1)
```

```python
# check Anova for columns_to_be_kept 

from scipy.stats import f_oneway

columns_to_be_kept_numerical = []

for i in columns_to_be_kept:
    a = list(df[i])  
    b = list(df['Approved_Flag'])  
    
    group_P1 = [value for value, group in zip(a, b) if group == 'P1']
    group_P2 = [value for value, group in zip(a, b) if group == 'P2']
    group_P3 = [value for value, group in zip(a, b) if group == 'P3']
    group_P4 = [value for value, group in zip(a, b) if group == 'P4']


    f_statistic, p_value = f_oneway(group_P1, group_P2, group_P3, group_P4)

    if p_value <= 0.05:
        columns_to_be_kept_numerical.append(i)
```

```python
# feature selection is done for cat and num features




# listing all the final features
features = columns_to_be_kept_numerical + ['MARITALSTATUS', 'EDUCATION', 'GENDER', 'last_prod_enq2', 'first_prod_enq2']
df = df[features + ['Approved_Flag']]
```

```python
import json
import pandas as pd
from joblib import load

# Load artifacts
MODEL_PATH = "models/xgb_classifier.joblib"
ENCODER_PATH = "models/label_encoder.joblib"
FEATURES_PATH = "models/feature_columns.json"

xgb_model = load(MODEL_PATH)
label_encoder = load(ENCODER_PATH)
with open(FEATURES_PATH, "r") as f:
    FEATURE_COLUMNS = json.load(f)

# Same EDUCATION mapping used during training
EDU_MAP = {
    "SSC": 1,
    "12TH": 2,
    "GRADUATE": 3,
    "UNDER GRADUATE": 3,
    "POST-GRADUATE": 4,
    "OTHERS": 1,
    "PROFESSIONAL": 3,
}

CAT_COLS = ["MARITALSTATUS", "GENDER", "last_prod_enq2", "first_prod_enq2"]

def preprocess(raw_df: pd.DataFrame) -> pd.DataFrame:
    df = raw_df.copy()

    # EDUCATION ordinal encode
    if "EDUCATION" in df.columns:
        df["EDUCATION"] = df["EDUCATION"].map(EDU_MAP).astype("Int64")

    # One-hot encode categorical columns exactly like training
    df = pd.get_dummies(df, columns=[c for c in CAT_COLS if c in df.columns])

    # Align to training feature columns (missing -> 0, extra -> dropped)
    df = df.reindex(columns=FEATURE_COLUMNS, fill_value=0)
    return df

def predict(df_raw: pd.DataFrame):
    X = preprocess(df_raw)
    y_pred_idx = xgb_model.predict(X)
    y_pred = label_encoder.inverse_transform(y_pred_idx)
    return y_pred

def predict_proba(df_raw: pd.DataFrame):
    X = preprocess(df_raw)
    return xgb_model.predict_proba(X)

if __name__ == "__main__":
    # Example: single applicant as dict
    sample = {
        # include all numeric features used in training plus these categorical fields:
        "EDUCATION": "GRADUATE",
        "MARITALSTATUS": "MARRIED",
        "GENDER": "M",
        "last_prod_enq2": "PL",       # example values
        "first_prod_enq2": "PL",      # example values
        # ... add other numeric feature columns expected by the model ...
    }
    df_sample = pd.DataFrame([sample])
    print("Predicted class:", predict(df_sample)[0])
    print("Class probabilities:", predict_proba(df_sample)[0])
```
